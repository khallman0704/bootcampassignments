```python
#inport
import pandas as pd

Location = "GleeClubData.csv"
df = pd.read_csv(Location)
```

```python
df.rename(columns={'# of practices per month': 'practices'}, inplace=True)
df.rename(columns={'# of competitions before regionals':'competitions'}, inplace=True)
df.rename(columns={'Regionals Score':'score'}, inplace=True)
df.head()
```

```python
df.head()
```

```python
df.corr()
```

```python
#df.Regionals Score
df.dropna(inplace = True)
df.head()
df["score"].where(df["score"] < 4.25, inplace=True)
```

```python
df.head()
```

```python
#performance practices per mo, team size, competitions = scores
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

plt.scatter(df['score'],df['practices'])
plt.xlabel('score')
plt.ylabel('practices')



#matplotlib.style.use('ggplot'
#                    pd.scatter_matrix(df,figsize=(10,10)
 #                                   ))
    
plt.show
```

```python
plt.scatter(df['score'],df['competitions'])
plt.xlabel('score')
plt.ylabel('competitions')
```

```python
import numpy as np

df.score = df.score.astype(np.int64)
df.practices= df.practices.astype(np.int64)
df.competitions = df.competitions.astype(np.int64)
```

```python
df.head()
```

```python
import statsmodels.formula.api as sm
result = sm.ols(formula='score ~ practices + competitions', data=df).fit()
result.summary()
```

```python
import statsmodels.formula.api as sm
result = sm.ols(formula='score ~ practices', data=df).fit()
result.summary()
```
